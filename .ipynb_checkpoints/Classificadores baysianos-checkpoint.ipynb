```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

blobs2 = pd.read_csv("blobs2.csv", names=["x1", "x2", "y"])
len_2 = blobs2.shape[0]

def training(dataset, tamanho):
    classes = {}
    # Separando as classes
    for i in range(0, tamanho):
        row = [x for x in np.array(dataset.iloc[[i]])[0]]
        if row[-1] not in classes:
            classes[row[-1]] = []
        classes[row[-1]].append(row)
    
    # Calculando média e covariancia
    mean_covariance_classes = {}

    for key in classes.keys():
        mean_covariance_classes[key] = {}
        for row in range(0, dataset.shape[0]):
            for col in range(0, dataset.shape[1] - 1):
                print(classes[key][row][col])
                #mean_covariance_classes[key][col] += classes[key][row][col]
                
            
    print(mean_covariance_classes)
            

training(blobs2, len_2)
```
