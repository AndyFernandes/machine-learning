# Redes neurais

```python
import numpy as np
import matplotlib.pyplot as plt
import common as com

import activation_functions
    
z = np.linspace(-10, 10, 100) 
    
fig, axs = plt.subplots(1, 3, sharex=True, sharey=False, figsize=(18,5))
plt.rcParams.update({'font.size': com.fontsize})

axs[0].plot(z, activation_functions.sigmoid(z), '-r', linewidth=com.linewidth, label='activation function')
axs[0].set_title("Sigmoid")

axs[1].plot(z, activation_functions.tanh(z), '-r', linewidth=com.linewidth, label='activation function')
axs[1].set_title("Tanh")

axs[2].plot(z, activation_functions.relu(z), '-r', linewidth=com.linewidth, label='activation function')
axs[2].set_title("ReLU")

plt.savefig('figs/activations.png', dpi=com.dpi)
```

```python
fig, axs = plt.subplots(1, 3, sharex=True, sharey=False, figsize=(18,5))
plt.rcParams.update({'font.size': com.fontsize})

axs[0].plot(z, activation_functions.grad_sigmoid(z), '-b', linewidth=com.linewidth, label='derivative')
axs[0].set_title("Sigmoid derivative")

axs[1].plot(z, activation_functions.grad_tanh(z), '-b', linewidth=com.linewidth, label='derivative')
axs[1].set_title("Tanh derivative")

axs[2].plot(z, activation_functions.grad_relu(z), '-b', linewidth=com.linewidth, label='derivative')
axs[2].set_title("ReLU derivative")

plt.savefig('figs/activations_derivatives.png', dpi=com.dpi)
```
