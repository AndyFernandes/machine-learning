# Redes neurais

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import neural_networks as neural_network
import common as com

from sklearn import datasets

np.random.seed(12345)

x = np.array([[0,0], [0,1], [1,0], [1,1]])
y = np.array([0,1,1,0])
labels = [0,1]
labels_text = ['F','V']
colors = ['red', 'blue']

def plot_data(x, y, marker='o', legend=True, only_data=False):
    plt.rcParams.update({'font.size': com.fontsize})
    for i, label in enumerate(labels):
        if only_data:
            plt.scatter(x[y == label,0], x[y == label,1],
                        s=10*com.markersize, marker=marker, color=colors[i], edgecolors= "black")
        else:        
            plt.scatter(x[y == label,0], x[y == label,1],
                        label=labels_text[i], s=10*com.markersize, marker=marker, color=colors[i], edgecolors= "black")    
    if only_data is False:
        if legend:
            plt.legend()
    
fig = plt.figure(figsize=com.figsize)
plot_data(x, y, legend=False)
plt.xlim(-0.5, 1.5)
plt.ylim(-0.5, 1.5)
#plt.savefig('figs/xor.png', dpi=com.dpi)
```

## Solução via Perceptron Simples

```python
model_perceptron = neural_network.train_perceptron(x=x, y=y.copy(), alpha=1, num_epochs=100)  
loss_history_perceptron = model_perceptron['loss_history']
loss_perceptron = loss_history_perceptron[-1]

if len(loss_history_perceptron) > 1:
    com.plot_loss_path(loss_history_perceptron)

plt.rcParams.update({'font.size': com.fontsize})
fig, ax = plt.subplots(figsize=com.figsize)
plt.xlim(-0.5, 1.5)
plt.ylim(-0.5, 1.5)
plot_data(x, y, legend=False)
xx, yy = com.make_meshgrid(x[:,0], x[:,1])
com.plot_contours(ax, model_perceptron,
                  lambda x, model: neural_network.linear_class_predic(model_perceptron, x),
                  xx, yy, colors)
#plt.savefig('figs/or_perceptron.png', dpi=com.dpi)
```
